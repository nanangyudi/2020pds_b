```
# Melakukan Import dari Library
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```
# Melakukan Pembacaan CSV Books
data = pd.read_csv("/content/drive/MyDrive/Poyek Akhir PDS/summary_data_sars.csv")
```

```
# menghapus variabel yang tidak perlu
drop_cols = ['Percentage of Imported cases', 'Number of HCW affected', 'Percentage of HCW affected']
data = data.drop(drop_cols, axis = 1)
```

```
#melakukan pengurutan data
data_sort = data.sort_values(by = 'No. of deaths', ascending=False).head()
data_sort.head()
```

# Pemodelan Data

```
#histogram kematian sars
num_bins=10
plt.hist(data['No. of deaths'], num_bins)
```

```
sns.distplot(data['No. of deaths'],bins=10)
```

```
num_bins=10
plt.hist(data['Cumulative total cases'],num_bins)
```

```
# hitung jumlah kematian
mean_dist = data.groupby('No. of deaths').size()
```

```
mean_dist
```

```
#Distribusi tabel mean age
mean_dist.plot(title='Jumlah Kematian')
```

```
cor = data.corr()
```

```
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns, annot=True)
```

```
#Menampilkan jumlah kematian di tiap negara
sns.set(rc={'figure.figsize':(30,20)})
sns.countplot(x='Country/Region', data=data)
```

```
sns.pairplot(data)
```

```
sns.relplot(x='No. of deaths',y='Cumulative total cases', hue='Country/Region',data=data)
```
