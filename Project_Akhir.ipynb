**Deskripsi data**

Data Nasional tentang frekuensi relatif dari nama-nama yang diberikan dalam populasi kelahiran A.S. di mana setiap individu memiliki Nomor Jaminan Sosial (ditabulasikan berdasarkan catatan Jaminan Sosial per 3 Maret 2019) Untuk setiap tahun kelahiran YYYY setelah 1879, kami membuat berkas yang dibatasi koma yang disebut yobYYYY.txt. Setiap catatan dalam berkas tahunan individu memiliki format "nama, jenis kelamin, nomor," di mana nama adalah 2 hingga 15 karakter, jenis kelamin adalah P (pria) atau W (wanita) dan "angka" adalah jumlah kemunculan nama tersebut. Setiap berkas diurutkan pertama berdasarkan jenis kelamin dan kemudian pada jumlah kejadian dalam urutan menurun. Ketika ada kesamaan pada jumlah kemunculan, nama-nama terdaftar dalam urutan abjad. Penyortiran ini memudahkan untuk menentukan peringkat nama. Catatan pertama untuk setiap jenis kelamin memiliki peringkat 1, catatan kedua untuk setiap jenis kelamin memiliki peringkat 2, dan seterusnya. Untuk menjaga privasi, kami membatasi daftar nama kami untuk mereka yang memiliki paling sedikit 5 kejadian.

Tujuannya untuk apakah bisa memprediksi suatu jenis kelamin seseorang dengan berbagai kombinasi nama.
Apakah bisa membedakan jenis kelamin dengan melihat dari nama?
